### 1. 15 points

Find the pdf for $Y = \tan(X)$ where $X$ is a uniform random variable over $(-\pi/2, \pi/2)$.

You should derive the form of the pdf by hand.

#### Solution

First we solve for X using an invertible transformation:
$Y = \tan(X)$ so $X = \arctan(Y) = h(Y)$

Note that tan(X) follows a Cauchy distribution[https://en.wikipedia.org/wiki/Cauchy_distribution], so from this, we can conclude that: 

$f_X(x) = \left\{ 
    \frac{1}{\pi}, \frac{-\pi}{2} \leq x \leq \frac{\pi}{2},
    0, \text{ otherwise} 
\right.$ 

$f_Y(y) = f_X(h(Y))|\frac{dh(y)}{dy}|$ 

$f_Y(y) = f_X(arctan(y))|\frac{dh(y)}{dy}|$ 

$f_Y(y) = \frac{1}{\pi}|\frac{dh(y)}{dy}|$ 

For the $|\frac{dh(y)}{dy}|$ statement, we will take the derivative of $arctan(y)$ in the following steps:


$\frac{dh(y)}{dy} arctan(y) = \frac{1}{\sec^2(y)} = \frac{1}{1 + y^2}$

Therefore, $f_Y(y) = f_X(h(Y))|\frac{dh(y)}{dy}| = \frac{1}{\pi} * \frac{1}{1 + y^2} = \frac{1}{\pi(1+y^2)}, \frac{1}{\pi(1+\frac{-\pi}{2}^2)} \leq y \leq \frac{1}{\pi(1+\frac{\pi}{2}^2)} $ 


### 2. 15 points

Two instruments are used to measure the height of an bridge. The error made by the less accurate instrument is normally distributed with mean 0 and standard deviation $0.6$ meters. The error made by the more accurate instrument is normally distributed with mean 0 and standard deviation $0.4$ meters.



Assuming the two measurements are independent random variables, what is the probability that their average value is within 0.5 meters of the height of the tower?



You should determine the form of the probability by hand, but then you can use Python to compute the numerical value.

#### Solution

**Define random variables for this example**

Let $X_1 \sim \mathscr{N}(\mu_1, \sigma_1^2) = \mathscr{N}(0, 0.6^2)$.

Let $X_2 \sim \mathscr{N}(\mu, \sigma^2) = \mathscr{N}(0, 0.4^2)$.

For this problem we want $P(-0.5 \leq x \leq 0.5)$

For summing two normal variables, we can use the formula $\mathscr{N}(\mu_1 + \mu_2), \sigma_1^2 + \sigma_2^2)$.

However, in this problem we need to take an average of both of the terms, so 

$E[X] = \frac{1}{n}*\sum_{i=1}^{n} X_i, i = 1,2, n = 2$ and

$V[X] = (\frac{1}{2})^2 (\text {or} \frac{1}{4})*\sum_{i=1}^{n} \sigma_i ^2, i = 1,2, n = 2$

Let $W = X_1 + X_2$

$W(\mu) = \frac{0 + 0}{2} = 0$

$W(\sigma^2) = \frac{0.6^2 + 0.4^2}{4} = \frac{1}{4} * (0.36 + 0.16) = \frac{1}{4} (0.52) = 0.13$. 

Therefore, 

$W(\sigma) = \sqrt{.13} \approx .36055$

We can use a Z-score to calculate our sample mean being 0.5 in the following equation: $Z = \frac{0.5-0}{.36055} \approx 1.387$ and $ZN = \frac{-0.5-0}{.36055} \approx -1.387 $

Since we want a probability within 0.5 meters, this will be $P(x \leq Z) = p(w) \approx 0.9172$ and $P(x \geq ZN) = p(w) \approx 0.0828$, so $P(ZN \leq x \leq Z) = 0.9172 - 0.0828 \approx 0.8345$

```python
numerator = (0.5-0)
numerator2 = (-0.5 - 0)
sigma_w = np.sqrt(0.13)
Z = numerator / sigma_w
ZN = numerator2 / sigma_w
PZ =  stats.norm.cdf(Z)
PZN =  stats.norm.cdf(ZN)
print(Z)
print(ZN)
print(PZ)
print(PZN)
print(PZ-PZN)
```
